First, let's load the data generated by `learning_analysis.py`.

```python
from collections import defaultdict
import numpy as np

class LearningCurvePoint:
    def __init__(self):
        self.num_analytics_cards = 0
        self.time_taken = 0
        self.num_correct = 0
        self.raw_incremental_gain = 0.0
        self.norm_incremental_gain = 0.0
        self.total_increments = 0.0
        self.sum_difficulty = 0.0
        self.sum_abilities = 0.0
        self.sum_abilities_weighted = 0.0
        self.sum_abilities_weights = 0.0
        self.num_all_cards = 0  # of cards, both assessment and non-assessment

npz = np.load("results.npz")

topic_curves = npz['topic_curves'][()]
agg_curve = npz['agg_curve'][()]

# we are assuming the control group is called 'control'
alternatives = topic_curves.keys()
alternatives.remove('control')
```

Now, compute some simple topic-level summary stats.

```python
def topic_summary(topic_name):
    """Given the name of a topic, compute some summary stats for that topic."""
    stats = defaultdict(float)
    for alt in alternatives:
        if topic_name not in topic_curves[alt]:
            continue
        topic_curve = topic_curves[alt][topic_name]
        for card in topic_curve:
            if card != 'num_users':
                stats['total_cards'] += topic_curve[card].num_analytics_cards
                stats['total_correct'] += topic_curve[card].num_correct
    return dict(stats)

topic_summaries = {}
for alt in alternatives:
    topic_summaries[alt] = {topic: topic_summary(topic) for topic in topic_curves[alt]}
```

If you like, you can print out topic summaries in a variety of ways:

```python
# topic names in alphabetical order
#print "\n".join(sorted(topic_summaries)) + "\n"

# top 10 topics by number of assessment cards
topic_total_cards = {}
for alt in alternatives:
    print "\nAlternative %s" % alt
    topic_total_cards[alt] = [(t, ts['total_cards']) for t, ts in topic_summaries[alt].iteritems() if 'total_cards' in ts]
    topic_total_cards[alt] = sorted(topic_total_cards[alt], key=lambda s: s[1], reverse=True)
    print "\n".join([str(t) for t in topic_total_cards[alt]][:10])
```

Here, we define some plotting routines that can be called for any topic

```python
from matplotlib import pyplot as plt
from matplotlib import figure
import matplotlib as mpl
from pylab import *

mpl.rcParams['axes.grid'] = True

num_analytics_cards = 60

def num_users(d):
    return d.itervalues(). next().num_all_cards

def engagement_curve(topic, relative=False):
    # again, we're expecting the control group to be called 'control'
    curve_true = {}
    for alt in alternatives:
        curve_true[alt] = topic_curves[alt][topic]
    curve_false = topic_curves['control'][topic]
    
    x = []; y_true = defaultdict(list); y_false = []; y = defaultdict(list)
       
    for card_num in range(num_analytics_cards):
        for alt in alternatives:
            if card_num not in curve_true[alt]:
                continue
        if card_num not in curve_false:
            continue
        
        y_false.append(curve_false[card_num].num_all_cards / float(num_users(curve_false)))
        for alt in alternatives:
            y_true[alt].append(curve_true[alt][card_num].num_all_cards / float(num_users(curve_true[alt])))
            y[alt].append(float(y_true[alt][-1]) / y_false[-1])
        x.append(card_num)
    
    if relative:
        for alt in alternatives:
            plt.plot(x, y[alt])
        plt.title("%s: relative engagement curve" % topic)
    else:
        for alt in alternatives:
            plt.plot(x, y[alt])
        plt.plot(x, y_false)
        plt.title("%s: engagement curve" % topic)

  
def learning_curves(topic, conditional=True, use_ability=False, aggregate=False, title=""):
    curve_true = {}
    for alt in alternatives:
        curve_true[alt] = topic_curves[alt][topic]
    curve_false = topic_curves['control'][topic]
    
    num_alts = len(alternatives)
    x = [];
    y_true = defaultdict(list); y_false = []; y_true_sum = defaultdict(int); y_false_sum = 0.
    
    user_scale = lambda card: card.num_all_cards if aggregate else 1.0
           
    for card_num in range(num_analytics_cards):
#        print curve_true[alt][card_num].num_analytics_cards
        error = False
        for alt in alternatives:
            if card_num not in curve_true[alt] or not curve_true[alt][card_num].num_analytics_cards:
                error = True
                break
        if error:
            continue
      
        if card_num not in curve_false or not curve_false[card_num].num_analytics_cards:
            continue
        if conditional:
            if use_ability:
                for alt in alternatives:
                    y_true_sum[alt] += curve_true[alt][card_num].norm_incremental_gain / curve_true[alt][card_num].total_increments * user_scale(curve_true[alt][card_num])
                    y_true[alt].append(y_true_sum[alt] / float(num_users(curve_true[alt])))
                y_false_sum += curve_false[card_num].norm_incremental_gain / curve_false[card_num].total_increments * user_scale(curve_false[card_num])
                y_false.append(y_false_sum / float(num_users(curve_false)))
            else:
                for alt in alternatives:
                    y_true_sum[alt] += curve_true[alt][card_num].raw_incremental_gain / curve_true[alt][card_num].total_increments * user_scale(curve_true[alt][card_num])
                    y_true[alt].append(y_true_sum[alt] / float(num_users(curve_true[alt])))
                y_false_sum += curve_false[card_num].raw_incremental_gain / curve_false[card_num].total_increments * user_scale(curve_false[card_num])
                y_false.append(y_false_sum / float(num_users(curve_false)))
        else:
            if use_ability:
                for alt in alternatives:
                    y_true[alt].append(curve_true[alt][card_num].sum_abilities / (curve_true[alt][card_num].num_analytics_cards))
                y_false.append(curve_false[card_num].sum_abilities / (curve_false[card_num].num_analytics_cards))
            else:
                for alt in alternatives:
                    y_true[alt].append(float(curve_true[alt][card_num].num_correct) / (curve_true[alt][card_num].num_analytics_cards))
                y_false.append(float(curve_false[card_num].num_correct) / (curve_false[card_num].num_analytics_cards))
            
        x.append(card_num)
    
    for alt in alternatives:
        plt.plot(x, y_true[alt])
    plt.plot(x, y_false)
    plt.title("%s: %s" % (topic, title))

    
```

Finally, study the gains in good ol' fashioned table form.

```python
def compute_gains(topic, alt, use_ability=False):
    curve = topic_curves[alt][topic]

    total_gain = 0.0
    total_increments = 0.0
    total_all_cards = 0.0
    
    for card_num, card in curve.iteritems():
        if hasattr(card,'raw_incremental_gain'):
            total_gain += card.norm_incremental_gain if use_ability else card.raw_incremental_gain
            total_increments += card.total_increments
            total_all_cards += card.num_all_cards
    return (total_gain / float(total_increments)) / float(num_users(curve)), total_all_cards

def compute_time(topic, alt, use_ability = False):
    curve = topic_curves[alt][topic]

    total_time = 0.0
    total_increments = 0.0
    total_all_cards = 0.0
    
    for card_num, card in curve.iteritems():
        if hasattr(card,'time_taken'):
            total_time += card.time_taken
            total_increments += card.total_increments
            total_all_cards += card.num_all_cards
    return (total_time / float(total_increments)) / float(num_users(curve)), total_all_cards


def compute_gains_table(start_topic, stop_topic, use_ability=True):
    
    sum_gains_true = defaultdict(int)
    sum_gains_false = 0.0
    sum_cards_true = defaultdict(int)
    sum_cards_false = 0.0
    gains_true = defaultdict(list)

    for (topic, _) in topic_total_cards[alternatives[0]][start_topic:stop_topic]:
        for alt in alternatives:
            gains_true[alt] = compute_gains(topic, alt, use_ability)
        gains_false = compute_gains(topic, 'control', use_ability)
        
        for alt in alternatives:
            sum_gains_true[alt] += gains_true[alt][0] * gains_true[alt][1]
        sum_gains_false += gains_false[0] * gains_false[1]
        
        for alt in alternatives:
            sum_cards_true[alt] += gains_true[alt][1]
        sum_cards_false += gains_false[1]

        print "Topic = %s" % topic
        for alt in alternatives:
            print "%s: %.4f\t%.2f" % (alt, gains_true[alt][0], gains_true[alt][0] * gains_true[alt][1])
        print "Control: %.4f\t%.2f\n" % (gains_false[0]*1000, gains_false[0] * gains_false[1])

    for alt in alternatives:
        print "\n%s" % alt
        print "Total gains: %s" % sum_gains_true[alt]
        print "Total number of cards: %s" % sum_cards_true[alt]
        print "Average gain per card: %s" % (sum_gains_true[alt] / sum_cards_true[alt])
        print "Relative change in agg gains: %.2f" % (sum_gains_true[alt] / sum_gains_false)
        print "Relative change in gain efficiency: %.4f" % ((sum_gains_true[alt] / sum_cards_true[alt]) / (sum_gains_false / sum_cards_false))
    
    print "Control"
    print "Total gains: %s" % sum_gains_false
    print "Total number of cards: %s" % sum_cards_false
    print "Average gain per card: %s" % (sum_gains_false / sum_cards_false)

def compute_time_table(start_topic, stop_topic, use_ability=True):
    
    sum_time_true = defaultdict(int)
    sum_time_false = 0.0
    sum_cards_true = defaultdict(int)
    sum_cards_false = 0.0
    time_true = defaultdict(list)

    for (topic, _) in topic_total_cards[alternatives[0]][start_topic:stop_topic]:
        for alt in alternatives:
            time_true[alt] = compute_time(topic, alt, use_ability)
        time_false = compute_time(topic, 'control', use_ability)
        
        for alt in alternatives:
            sum_time_true[alt] += time_true[alt][0] * time_true[alt][1]
        sum_time_false += time_false[0] * time_false[1]
        
        for alt in alternatives:
            sum_cards_true[alt] += time_true[alt][1]
        sum_cards_false += time_false[1]

        print "Topic = %s" % topic
        for alt in alternatives:
            print "%s: %.4f\t%.2f" % (alt, time_true[alt][0], time_true[alt][0] * time_true[alt][1])
        print "Control: %.4f\t%.2f" % (time_false[0]*1000, time_false[0] * time_false[1])

    for alt in alternatives:
        print "\n%s" % alt
        print "Total time: %s" % sum_time_true[alt]
        print "Total number of cards: %s" % sum_cards_true[alt]
        print "Average time per card: %s" % (sum_time_true[alt] / sum_cards_true[alt])
        print "Relative change in time spent: %.2f" % (sum_time_true[alt] / sum_time_false)
        print "Relative change in average time spent: %.4f" % ((sum_time_true[alt] / sum_cards_true[alt]) / (sum_time_false / sum_cards_false))
    
    print "Control"
    print "Total gains: %s" % sum_time_false
    print "Total number of cards: %s" % sum_cards_false
    print "Average time spent per card: %s" % (sum_time_false / sum_cards_false)
    
compute_time_table(0, 10, False)
```
